```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import pyplot as bar
from scipy.stats import multivariate_normal as mvn
```

## Importing Data

```python
data = pd.read_csv('https://raw.githubusercontent.com/VIJAY07102/ITIT-4103-2021/main/Assignment%205/iris.data', header=None).values
np.random.shuffle(data)
X = data[:, :4]
Y = data[:, 4:]
feature=['SepalLength','SepalWidth','PetalLength','PetalWidth']
classes=['Iris-setosa','Iris-versicolor','Iris-virginica']
```

## Bayes Classifier

```python
class NaiveBayes(object):
    def fit(self, X, Y, smoothing=10e-4):
        self.gaussians = dict()
        self.priors = dict()
        labels = set(Y)
        for c in labels:
            current_x = X[Y == c]

            self.gaussians[c] = {
                'mean': current_x.mean(axis=0),
                'var': current_x.var(axis=0) + smoothing,
            }
            self.priors[c] = float(len(Y[Y == c])) / len(Y)

    def predict(self, X):
        N, D = X.shape
        K = len(self.gaussians)
        P = np.zeros((N, K))
        for c, g in self.gaussians.items():
            mean, var = g['mean'], g['var']
            P[:, c] = mvn.logpdf(X, mean=mean, cov=var,allow_singular=True) + np.log(self.priors[c])

        return np.argmax(P, axis=1)
```

### The main objective of this task is to find the most important features for accurately predicting the class of an input flower. In order to do we use a technique called forward feature selection. This reduction in features offers the following benefits
1. Reduced training times
2. Simplified and interpretable models
3. Reduced chances of overfitting i.e. lesser variance
4. Less impact of the curse of dimensionality
5. Forward feature selection allows us to tune this hyperparameter for optimal performance.

#### We can observe that having selected 3 or 4 features we get the better accuracy, and the accuracy decreases as we reduce the number of features selected for training and predicting. Since, it's a small dataset with less number of features, we are able to use forward feature selection with all possible ways and can get the best features to find the good discriminants.
